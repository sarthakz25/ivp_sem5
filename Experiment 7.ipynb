```python
# imports

from skimage import io
from scipy import signal
from skimage.color import rgb2gray
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image= cv2.imread('morph.png', 0)
```

```python
sh=image.shape
r=sh[0]
c=sh[1]

for row in range(0,r):
    for col in range(0,c):
        if image[row][col]>100:
           image[row][col]=255
        else:
           image[row][col]=0
```

```python
plt.imshow(image, cmap='gray')
```

```python
# erosion

image_erosion=image.copy()
sz=11
se=255*(np.ones([sz,sz]))

cent=int((sz-1)/2)
sh=image_erosion.shape
r=sh[0]
c=sh[1]
```

```python
for row in range(cent,r):
    for col in range(cent,c):
        temp=image[row-cent:row+cent+1,col-cent:col+cent+1]
        if np.array_equal(temp,se):
           image_erosion[row][col]=255
        else:
           image_erosion[row][col]=0
```

```python
plt.figure(figsize=(10,10))

plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1,2,2)
plt.imshow(image_erosion, cmap='gray')
plt.title('Eroded Image')
```

```python
# dilation

image_dilation=image.copy()
sz=11
se=255*(np.ones([sz,sz]))
cent=int((sz-1)/2)
sh=image_dilation.shape
r=sh[0]
c=sh[1]
```

```python
for row in range(cent,r):
    for col in range(cent,c):
        temp=image[row-cent:row+cent+1,col-cent:col+cent+1]
        if np.isin(255,temp):
           image_dilation[row][col]=255
        else:
           image_dilation[row][col]=0
```

```python
plt.figure(figsize=(10,10))

plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1,2,2)
plt.imshow(image_dilation, cmap='gray')
plt.title('Dilated Image')
```

```python
# Image opening

# Erosion

image_erosion=image.copy()
sz=11
se=255*(np.ones([sz,sz]))

cent=int((sz-1)/2)
sh=image_erosion.shape
r=sh[0]
c=sh[1]

for row in range(cent,r):
    for col in range(cent,c):
        temp=image[row-cent:row+cent+1,col-cent:col+cent+1]
        if np.array_equal(temp,se):
           image_erosion[row][col]=255
        else:
           image_erosion[row][col]=0

# dilation of eroded image

image_dilation=image.copy()

for row in range(cent,r):
    for col in range(cent,c):
        temp=image_erosion[row-cent:row+cent+1,col-cent:col+cent+1]
        if np.isin(255,temp):
           image_dilation[row][col]=255
        else:
           image_dilation[row][col]=0
```

```python
plt.figure(figsize=(10,10))

plt.subplot(1,3,1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1,3,2)
plt.imshow(image_erosion, cmap='gray')
plt.title('Eroded Image')

plt.subplot(1,3,3)
plt.imshow(image_dilation, cmap='gray')
plt.title('Dilation of eroded Image')
```
