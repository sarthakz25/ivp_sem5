```python
import matplotlib.image as mimg
import matplotlib.pyplot as plt
```

```python
image = mimg.imread('ivp-images_exp1\cat.png')
print(image.shape, image.dtype, type(image))
```

```python
image[100,100]
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
image1 = image.copy()
image2 = image.copy()

# to change image to darker color
image1[image1<0.75] = 0
# to change image to lighter color
image2[image2>0.25] = 1

plt.subplot(2,2,1)
plt.imshow(image)
plt.axis('off')
plt.title('original')

plt.subplot(2,2,4)
plt.imshow(image2)
plt.axis('off')
plt.title("lightened")

plt.subplot(2,2,3)
plt.imshow(image1)
plt.axis('off')
plt.title('darkened')
plt.savefig('ivp-images_exp1\cat_dark.png')
img=mimg.imread('ivp-images_exp1\cat_dark.png')

plt.show()
```

```python
from skimage.io import imread, imsave, imshow
from skimage import color
```

```python
im3 = imread('ivp-images_exp1\ishihara.jpg')
im4 = color.rgb2gray(im3)

plt.figure(figsize=(10,10))

plt.subplot(1,2,1)
plt.imshow(im3)
plt.title('Original Color Image')

plt.subplot(1,2,2)
plt.imshow(im4, cmap='gray')
plt.title('Converted Gray Image')
```

```python
from skimage.transform import resize
import numpy as np

# dividing by 255 to normalize
im5 = mimg.imread('ivp-images_exp1\dhoni.jpg')/255
im6 = mimg.imread('ivp-images_exp1\kolhi.jpeg')/255

sh=im6.shape

# images should be of same size
# numpy resize does not work axis wise
im7 = resize(im7,(sh[0],sh[1],3))

plt.figure(figsize=(20,15))

i=1
for alpha in np.linspace(0,1,20):
  plt.subplot(4,5,i)
  imm=((1-alpha)*im7)+(alpha*im6)
  plt.imshow(imm)
  i=i+1
  plt.subplots_adjust(wspace=0.05,hspace=0.05)
```
