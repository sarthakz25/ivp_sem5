```python
# imports
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from skimage.io import imread, imsave, imshow
from skimage import color
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# loading
img1= cv2.imread("cameraman.png",0)
#type(img1)
print(np.amax(img1))
plt.imshow(img1)
```

```python
plt.imshow(img1,cmap="gray")
```

```python
[m,n]=img1.shape
print(m,n)
```

```python
b = int(input("Enter a value between 1 and 7: "))
const = (2**b - 1)/255
print(const)
```

```python
# image compression

i = 1
while i < 9:
  b = i
  const = (2**b - 1) / 255
  print(const)

  img2 = np.zeros((m, n))

  for row in range(m):
    for col in range(n):
      img2[row, col] = const * img1[row, col]
  img2 = np.round(img2)

  plt.subplot(4, 2, i)
  plt.title("b =" + str(b))
  plt.imshow(img2, cmap='gray')
  plt.axis('off')

  cv2.imwrite("grey" + str(i) + ".png", img2)

  i += 1

plt.show()
```
